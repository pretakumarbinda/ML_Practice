```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 1000)  
sin_wave = np.cos(2 * np.pi * 5 * t)
plt.plot(t, sin_wave) 
plt.title("Cosine Wave")
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 1000)  
sin_wave = np.sin(2 * np.pi * 5 * t)
plt.plot(t, sin_wave) 
plt.title("Sine Wave")
plt.grid()
plt.show()
```
