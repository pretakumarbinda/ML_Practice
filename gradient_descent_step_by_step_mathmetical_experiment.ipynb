```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
plt.scatter(X,y)
```

```python
# let's first see the co-efficients (value of m and b)
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X,y)
```

```python
print(reg.coef_)          #value of m
print(reg.intercept_)     #value of b
```

```python
 # Now we will find this value of b by using the value of m=78.35 using Gradient Decent

# let b=0
y_pred = (78.35*X + 0)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X, y_pred, color='green', label='b=0')
plt.legend()
plt.show()
```

```python
# let's see the slope:
m=78.35
b=0
loss_slope = -2*np.sum(y -m*X.ravel() - b)
print(loss_slope)
```

```python
# now if we let learning rate = 0.1
lr=0.1
step_size=lr*loss_slope
step_size
```

```python
#now b_new will be:
b = b - step_size
b
```

```python
y_pred1 = (m*X+b).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#A3E4D7', label='b=0')
plt.plot(X,y_pred1, color='#00a65a', label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
#iteration 02:
loss_slope = -2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size = lr*loss_slope
b = b-step_size
b
```

```python
y_pred2 = (m*X+b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color='red', label='OLS')
plt.plot(X, y_pred, color='#A3E4D7', label='b=0')
plt.plot(X,y_pred1, color='#00a65a', label='b={}'.format(b)) #ignore this green label
plt.plot(X,y_pred2, color='black', label='b={}'.format(b))
plt.legend()
plt.show()
```

```python
# like this, we are getting the right value of b
```

```python
# but we can do it also using iterations

b=-100
m=78.35
lr=0.1
epochs =10

for i in range(epochs):
    loss_slope =-2*np.sum(y-m*X.ravel()-b)
    b = b - lr*loss_slope
    y_pred = m*X+b

    plt.plot(X,y_pred)

plt.scatter(X,y)
```

```python
class GDRegressor:

    def __init__(self, learning_rate, epochs):
        self.m = 61.15548258
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs
    def fit(self, X,y):
        for i in range(self.epochs):
            loss_slope = -2*np.sum(y-self.m*X.ravel() - self.b)
            self.b = self.b - (self.lr*loss_slope)
        print(self.b)
```
